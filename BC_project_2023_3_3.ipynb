```python
#@title Simulation Demo { display-mode: "both" }

import matplotlib.pyplot as plt

# set initial and k values
p0 = 0.4 #@param {type:"number"}
k = 3.8 #@param {type:"number"}
generations = 100 #@param {type:"integer"}


def calc(prev,k):
    return (k*prev*(1-prev))

prev = p0
print("0th generation: ",p0, " ,k = ",k)

results = []

for i in range(1,generations+1):
    current = calc(prev,k)
    #print(i,"th generation: ",current)
    results.append(current)
    prev = current

plt.plot(results)
plt.title("Logistics Sequence Graph")
plt.figtext(1,1,"k value: "+str(k))
plt.figtext(1,1.05,"p0 value: "+str(p0))
plt.xlabel("ith Generation")
plt.ylabel("Population")
plt.grid()

plt.show()
```
